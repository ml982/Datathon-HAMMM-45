```python
import pandas as pd
```

```python
df = pd.read_csv('RoughSleepers.csv')
```

```python
df = df.set_index(list(df)[0])
```

```python
df 
```

```python
df = df.drop('GSS Code', axis=1)
```

```python
df = df.drop(['2019/20','2020/21','2021/22'],axis=1)
```

```python
df
```

```python
df.iloc[0, 1]
```

```python
df['proportion'] = df['2023/24'] / df.iloc[0, 1]
```

```python
df
```

```python
df2 = df.drop('Greater London Authority')
plot = df2.plot.pie(y='proportion', figsize=(15, 15))
```

```python
a = 0.4
```

```python
totalbudget = 100000
```

```python
b = 0.6
```

```python
df['food funding'] = df['proportion'] * a * totalbudget
```

```python
df['shelter funding'] = df['proportion'] * b * totalbudget
```

```python
df['Funding_Needed'] = df['shelter funding'] + df['food funding']
```
